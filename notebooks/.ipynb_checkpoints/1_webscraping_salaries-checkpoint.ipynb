NOW LETS LOOP THROUGH ALL PAGES

```python
from urllib.request import urlopen
```

```python
web_pages = ['https://thesportsdaily.com/2010/12/29/2010-year-in-review-ufc-salaries-fox11/', 'https://thesportsdaily.com/2012/01/11/2011-year-in-review-ufc-fighter-salaries-fox11-fox11/','https://thesportsdaily.com/2013/01/11/2012-year-in-review-ufc-fighter-salaries-fox11/','https://thesportsdaily.com/2013/01/11/2012-year-in-review-ufc-fighter-salaries-fox11/','https://thesportsdaily.com/2014/12/27/2014-ufc-fighter-salaries-fox11/','https://thesportsdaily.com/2016/01/05/2015-ufc-fighter-salaries-fox11/','https://thesportsdaily.com/2017/01/03/2016-ufc-fighter-salaries-fox11/','https://thesportsdaily.com/2018/01/06/2017-ufc-fighter-salaries-complete-list-fox11/','https://thesportsdaily.com/2019/01/13/2018-ufc-fighter-salaries-complete-list-fox11/','https://thesportsdaily.com/2019/12/30/2019-ufc-fighter-salaries-complete-list-fox11/amp/','https://thesportsdaily.com/2020/12/31/2020-ufc-fighter-salaries-complete-list-fox11/']
```

```python
salary_df = pd.DataFrame()

for i in web_pages:
    get_info = requests.get(i)
    now_soup = BeautifulSoup(get_info.content, 'lxml')
    cards = now_soup.find_all('td')
    all_data = []
    for stuff in cards:
        all_data.append(stuff.text)
    all_fighters= all_data[1::3]
    all_salaries = all_data[2::3]
    all_salaries = [s.replace("\xa0", "") for s in all_salaries]
    fighters_salaries = pd.DataFrame(list(zip(all_fighters, all_salaries)),
               columns =['fighter', 'salary'])
    salary_df = pd.concat([salary_df, fighters_salaries])

```

```python
salary_df
```

```python
salary_df.to_csv('../data/salary_df.csv')
```
