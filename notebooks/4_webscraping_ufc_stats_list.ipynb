```python
web_pages
```

```python
fighter_bio_df = pd.DataFrame()

for i in web_pages:
    get_info = requests.get(i)
    now_soup = BeautifulSoup(get_info.content, 'lxml')
    all_data = now_soup.find_all('td')
    data = []
    for s in all_data:
        data.append(s.text.strip())
    first_name = data[1::11]
    last_name = data[2::11]
    height = data[4::11]
    weight = data[5::11]
    weight= [h.replace(' lbs.', "") for h in weight]
    stance = data[7::11]
    wins = data[8::11]
    losses = data[9::11]
    draws = data[10::11]
    fighter_bio = pd.DataFrame(list(zip(first_name, last_name, height, weight, stance, wins, losses, draws)),
               columns =['first_name', 'last_name', 'height', 'weight', 'stance', 'wins', 'losses', 'draws'])
    fighter_bio_df = pd.concat([fighter_bio_df, fighter_bio])
```

```python
fighter_bio_df
```

```python
fighter_bio_df['wins'] = pd.to_numeric(fighter_bio_df['wins'])
```

```python
fighter_bio_df['losses'] = pd.to_numeric(fighter_bio_df['losses'])
```

```python
fighter_bio_df['draws'] = pd.to_numeric(fighter_bio_df['draws'])
```

```python
fighter_bio_df['number_of_fights'] = fighter_bio_df['wins'] + fighter_bio_df['losses']+fighter_bio_df['draws']
```

```python
fighter_bio_df['full_name'] = fighter_bio_df['first_name']+ ' ' + fighter_bio_df['last_name']
```

```python
fighter_bio_df
```

```python
fighter_bio_df.to_csv('../data/fighter_bio_df.csv')
```
